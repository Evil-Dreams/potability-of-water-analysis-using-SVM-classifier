```python
import kagglehub
from kagglehub import KaggleDatasetAdapter as kda
data=kagglehub.dataset_load(kda.PANDAS,"vibishananandhan/water-potability","water_potability.csv")
```

```python
import pandas as pd
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
data['ph'].fillna(data['ph'].median(), inplace=True)
data['Sulfate'].fillna(data['Sulfate'].median(), inplace=True)
data['Trihalomethanes'].fillna(data['Trihalomethanes'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
x = data.drop('Potability', axis=1)
y = data['Potability']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svc_rbf = SVC(kernel='rbf',probability=True)
svc_rbf.fit(x_train, y_train)
```
